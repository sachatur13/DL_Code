```python
from tensorflow.keras.datasets import mnist
```

```python
(x_train,y_train),(x_test,y_test) = mnist.load_data()
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
## Visualizing using opencv
for i in range(0,7):
    rand = np.random.randint(0,len(x_train))
    image = x_train[rand]
    window = 'Random #' + str(i)
    cv2.imshow(window,image)
    cv2.waitKey(0)
cv2.destroyAllWindows()    
```

```python
## Reshaping data
x_train = x_train.reshape(x_train.shape[0],x_train[0].shape[0],x_train[1].shape[0],1)
x_test = x_test.reshape(x_test.shape[0],x_test[0].shape[0],x_test[1].shape[0],1)
```

```python
input_shape = (28,28,1)
x_train = x_train.astype('float32')
x_test = x_test.astype('float32')
x_train/=255
x_test/=255
```

```python
## One hot encoding the data
from tensorflow.keras.utils import to_categorical

y_train = to_categorical(y_train)
y_test = to_categorical(y_test)

n_classes = y_test.shape[1]
n_pixels = x_train.shape[1] * x_train.shape[2]
```

```python
## Building a model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D,MaxPooling2D,Flatten,Dropout
from tensorflow.keras.optimizers import SGD
```

```python
model = Sequential()
model.add(Conv2D(32,kernel_size=(3,3), activation='relu',input_shape = input_shape))
model.add(Conv2D(64,kernel_size = (3,3),activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.25))
model.add(Flatten())
model.add(Dense(128,activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(n_classes,activation='softmax'))

model.compile(loss = 'categorical_crossentropy',optimizer = 'adam',metrics = ['accuracy'])
print(model.summary)
```

```python
print(model.summary())
```

```python
## Model Training
batch = 32
epochs = 1

history = model.fit(x_train,y_train,batch_size = batch,epochs = epochs,verbose = False
,validation_data= (x_test,y_test))
```

```python
score = model.evaluate(x_test,y_test,verbose = False)
print(score)
```
