```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

#### Algos - sobel, laplacian, canny (best due to low error)

#### Canny ----
#### Gaussian blur -> find intensity gradient -> apply non max suppresion (remove non edges) -> apply thresholds to find edges

```python
input_image = cv2.imread(r'kim.jpg',0)
```

```python
plt.imshow(input_image)
```

```python
height,width = input_image.shape
```

```python
## sobel
sobel_x = cv2.Sobel(input_image,cv2.CV_64F,0,1,ksize=5)
sobel_y = cv2.Sobel(input_image,cv2.CV_64F,1,0,ksize=5)

plt.imshow(input_image);
```

```python
plt.imshow(sobel_x);
```

```python
plt.imshow(sobel_y);
```

```python
## Sobel_OR
sobel_or = cv2.bitwise_or(sobel_x,sobel_y)
plt.imshow(sobel_or)
```

```python
## Laplacian
lap = cv2.Laplacian(input_image,cv2.CV_16UC2)
plt.imshow(lap)
```

```python
## Canny
canny = cv2.Canny(input_image,150,150)
plt.imshow(canny)
```
