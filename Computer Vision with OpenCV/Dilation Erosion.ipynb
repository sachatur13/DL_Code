```python
## Dilation - adding pixels to the edges
## Erosion - removing pixels from the edges
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
input_image = cv2.imread(r'numbers.jpg')
```

```python
plt.imshow(input_image)
```

```python
kernel = np.ones((5,3),np.uint8)
##erosion
er = cv2.erode(input_image,kernel,iterations = 2)
plt.imshow(er)
```

```python
## dilation
dl = cv2.dilate(input_image,kernel,iterations = 2)
plt.imshow(dl)
```

```python
## opening
open_ = cv2.morphologyEx(input_image,cv2.MORPH_OPEN,kernel)
plt.imshow(open_)
```

```python
## Closing
close_ = cv2.morphologyEx(input_image,cv2.MORPH_CLOSE,kernel)
plt.imshow(close_)
```
