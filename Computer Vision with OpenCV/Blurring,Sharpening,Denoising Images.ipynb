```python
## This is a kernel operation
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
input_image = cv2.imread(r'elephant.jpg')
```

```python
plt.imshow(input_image)
```

```python
kernel = np.ones((3,3),np.float32)/9 ## 3x3 kernel

blurred_image = cv2.filter2D(input_image,-10,kernel)
plt.imshow(blurred_image)
```

```python
blur = cv2.medianBlur(input_image,5)
cv2.imshow('blur',blur)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
## Denoizing
dst = cv2.fastNlMeansDenoisingColored(input_image,None,10,6,7,32)
plt.imshow(dst)
```

```python
pyramid = cv2.imread(r'IMG_7539.jpg')
```

```python
sharp = np.array([[-1,-1,-1],
                 [-1,9,1],
                 [-1,-1,-1]]) 
sharp_image = cv2.filter2D(pyramid,-1,sharp)
plt.figure(figsize = (12,8))
plt.imshow(sharp_image)
```
