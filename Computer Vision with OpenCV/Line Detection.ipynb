```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
input_image = cv2.imread(r'soduku.jpg')
```

```python
plt.imshow(input_image)
```

```python
gray = cv2.cvtColor(input_image,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray)
```

```python
ret,thresh = cv2.threshold(gray,127,255,0)
plt.imshow(thresh)
```

```python
## find edges using canny
edge = cv2.Canny(gray,15,150)
plt.imshow(edge)
```

```python
## finding lines using HoughLines
lines = cv2.HoughLines(edge,1,np.pi / 180,150)

## Iterating through lines
for line in lines:
    rho,theta = line[0]
    a = np.cos(theta)
    b = np.sin(theta)
    x0 = a*rho
    y0 = b*rho
    
    ## Finding start and end points of the lines
    x1 = int(x0 + 1000 *(-b))
    y1 = int(y0 + 1000 *(a))
    x2 = int(x0 - 1000 *(-b))
    y2 = int(x0 - 1000 *(a))
    cv2.line(input_image,(x1,y1),(x2,y2),(120,0,0),1)

plt.figure(figsize = (10,12))
plt.imshow(input_image)
```

```python
input_image = cv2.imread(r'soduku.jpg')
```

```python
gray = cv2.cvtColor(input_image,cv2.COLOR_BGR2GRAY)
```

```python
egde =  cv2.Canny(gray,75,150)
```

```python
plt.imshow(egde)
```

```python
## Probabilistic Houghlines
lines = cv2.HoughLinesP(egde,1,np.pi/180,30,maxLineGap = 250)

for line in lines:
    x1,y1,x2,y2 = line[0]
    cv2.line(input_image,(x1,y1),(x2,y2),(0,0,128),1)
plt.imshow(input_image)
```
