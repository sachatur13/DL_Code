```python
import numpy 
import cv2
import matplotlib.pyplot as plt

input_image = cv2.imread(r'crowdyale.jpg')
```

```python
plt.imshow(input_image)
```

```python
to_find = cv2.imread(r'Capture.JPG',0)
```

```python
plt.imshow(to_find)
```

```python
## Converting to grayscale
gray = cv2.cvtColor(input_image,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray)
```

```python
result = cv2.matchTemplate(gray,to_find,cv2.TM_CCOEFF)
min_val,max_val,min_loc,max_loc = cv2.minMaxLoc(result)
```

```python
## creating a bounding box
top_left = max_loc
bottom_right = (top_left[0]+150,top_left[1]+150)
cv2.rectangle(input_image,top_left,bottom_right,(0,0,225),5)
plt.imshow(input_image)
```
