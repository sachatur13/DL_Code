### Visualizing the data

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (20,5))
data_copy.spx.plot()
plt.title('S&P Prices')
```

```python
plt.figure(figsize = (20,5))
data_copy.ftse.plot()
data_copy.spx.plot()
plt.title('FTSE Prices')
```

### QQ Plots

```python
import scipy.stats as st
import pylab
```

```python
plt.figure(figsize = (10,5))
st.probplot(data_copy.spx,plot = pylab)
plt.ylabel('SPX values')
plt.xlabel('Deviation from the mean')
```
